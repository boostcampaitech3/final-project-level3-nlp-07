```python
# 인스타 -> #@전화번호# 마스킹
print('-- ', df.iloc[30944]['사장답글'])
print('-- ', re.sub("\@[a-zA-Z0-9_]+", "#@전화번호#", df.iloc[30944]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("\@[a-zA-Z0-9_]+", "#@전화번호#")
```

```python
# 전화번호 -> #@전화번호# 마스킹
print('-- ', df.iloc[204910]['사장답글'])
print('-- ', re.sub("(2\.[\w\.]+)|(010-[\w-]+)", "#@전화번호#", df.iloc[204910]['사장답글']))

df["사장답글"] = df["사장답글"].str.replace("(2\.[\w\.]+)|(010-[\w-]+)", "#@전화번호#")
```

```python
# 이메일 및 홈페이지 (.co.kr 없음, naver/ daum 없음)
# - https://instagram.com/nkjnnk.bd/
# - jbhbk@gmail.com
df["사장답글"] = df["사장답글"].str.replace("[\w\/\:\.\@]+.com[\w\/\.]{0,}", "#@전화번호#")
```
